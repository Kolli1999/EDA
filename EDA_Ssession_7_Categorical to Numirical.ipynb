**Task**

- read the data again
- apply for loop
- iterate through columns
- Get the unique labels of each column
- create a dictinory
- apply in to the map

```python
cols=visa_df.select_dtypes(include="object").columns
for i in cols:
    labels=list(visa_df[i].unique())
    values=[i for i in range(len(labels))]
    d=dict(zip(labels,values))
    visa_df[i]=visa_df[i].map(d)
visa_df
```

**One hot encoder**

- one hot means if one is ON another one is OFF

- ON represents with 1, and OFF represents with 0

- For example case status has two unique labels

    - denied
    - certified

- if we apply one hot encoder on case status column it will create two extra columns

    - case_status_denied
    - case_status_Certified

|case_status|csae_status_denied|case_satus_certified
|-------|--|--|
|denied|1|0|
|certified|0|1

**Advantages**

- In machine learning we have one important property is **Independent of variables**

- Machine learning models expects variables are **independent each other**

- one hot coder creating new columns, the columns are **independent each other**

- Independent variable means both variables have **90 degrees phase shift**

- 90 degrees phase shift also called as **perepndicular each other**

- perpendicular menas **orthogonal each other**

**Disadvantages**

- One hot encoder creates new columns equlas to number of unique labels

- If number of unique labels increase, new columns also increased

- The more columns leads data is more

- Then data is more the complexity will increases

- Complexity increases more computation power required

- Also time also increase

- The entire process names as **curse of dimenionality**
