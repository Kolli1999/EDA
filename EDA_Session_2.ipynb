**Create data frame using dictionary**

```python
names=['Avinash','Akash','Aadhya']
age=[25,35,54]
pd.DataFrame(zip(names,age),columns=['nmaes','age'])
```

```python
dict1={'names':['Avinash','Akash','Aadhya'],'Ages':[25,35,54]}

pd.DataFrame(dict1,index=[1,2,3])
#Keys becomes columns
#values becoms rows
#No need to provise columns names sepeartly
```
