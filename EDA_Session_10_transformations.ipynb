**Data transformations**

- Generally data has 3 types of distributions

    - Normal distribution

    - postive skewed

    - Negative skewed

- All the maths concepts developed by assumptions as data follows data distribution

- Data skew happens because of many reasons, in that one reason is outliers

- Once we perform the outlier analysis then we can apply transformation methods, if the data doesn't follow normal distribution

- We have some transformation methods mentioned below

    - log transformation

    - Exponential transformation

    - Reciprocal transformation
 
    - square root transformation

    - power transformation

```python
#step-1
import numpy as np
import matplotlib.pyplot as plt
```

```python
#step-2:Read the data
exp_data=np.random.exponential(size=10000)
exp_data
```

```python
#step-3: plot the data
plt.hist(exp_data,bins=20,label='Exponential')
plt.show()
```

```python
norm_data=np.random.normal(size=10000)
norm_data
```

```python
plt.hist(norm_data,bins=20,label='Exponential')
plt.show()
```

**Log transformation**

- Log transformation means performing lgorithm operations on original data

- It is an approach to convert into Normal distribution

- Log menas natural logarithm base=e

```python
x=2
np.log(x)
```

```python
log_data=np.log(exp_data)
plt.hist(log_data,bins=20,label='Logarithm')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1).hist(exp_data,bins=30,label='Exponential')
plt.legend()
plt.subplot(1,2,2).hist(log_data,bins=20,label='Logarithm')
plt.legend()
plt.show()
```

**Reciprocal Transformations**

```python
x=2
1/x
```

- Reciprocal will fail when values are low

- for example x=0 then 1/0 fail

```python
resiprocal_data=np.reciprocal(exp_data)
plt.hist(resiprocal_data,bins=20,label='Logarithm')
plt.show()
```

$Square$ $root$ $transformation$

```python
square_root_data=np.sqrt(exp_data)
plt.hist(square_root_data,bins=20,label='Square Root Data')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.subplot(1,3,1).hist(exp_data,bins=20,label='Exponential')
plt.legend()
plt.subplot(1,3,2).hist(log_data,bins=20,label='Logarithm')
plt.legend()
plt.show()
plt.subplot(1,3,3).hist(square_root_data,bins=20,label='square')
plt.legend()
plt.show()
```
